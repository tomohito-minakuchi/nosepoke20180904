```julia
# Problem 2: x^y
function my_power(x, y)
    ans = 1
    for i in (1:y)
       ans *= x
    end
    return ans
end
my_power(4,3)
```

```julia
# Problem 3
function factorial_recursive(x)
    if x>1
        ans = x*factorial_recursive(x-1)
        else 
        ans = 1
    end
    return ans
end
factorial_recursive(6)
```

```julia
# Problem 3-2

function primenumber(x)
    ans = Int64[]
    for i = (2:x)
        temp = 1
        for j = ans
            temp *= i.%j
        end
        if temp != 0
            ans = append!(ans, i)
        end
    end
    return ans
end

exm = primenumber(10000)
```

```julia
# Problem 4: load files and print a string
using JLD
d = load("joyce.jld")
joyce = d["joyce"]

println(joyce)
```

```julia
function vowelcount(x)
counter = 0
    for c in joyce
        if c == x
            counter += 1
        end
    end
    return counter
end

vowelcount('e')
```
